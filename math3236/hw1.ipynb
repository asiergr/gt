```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
ten_sample = [np.random.exponential(scale=0.5, size=10) for _ in range(200)]
twenty_sample = [np.random.exponential(scale=0.5, size=10) for _ in range(200)]
fifty_sample = [np.random.exponential(scale=0.5, size=10) for _ in range(200)]
hundred_sample = [np.random.exponential(scale=0.5, size=10) for _ in range(200)]
samples = [ten_sample, twenty_sample, fifty_sample, hundred_sample]
```

```python
def ecdf(data):
    erange_x, counts = np.unique(data, return_counts=True)
    cdf_emp = np.cumsum(counts)/len(data)
    return cdf_emp, erange_x
```

```python
for sample in samples:
    sample_mean = [np.mean(subsample) for subsample in sample]
    data, counts = ecdf(sample_mean)
    plt.step(counts, data, linestyle='--', label='Empirical cdf')
    plt.plot(counts, norm.cdf(counts, loc=0.5, scale=0.25))
    plt.legend()
    plt.show()
```
