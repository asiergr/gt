```python
import numpy as np
from typing import Callable, Tuple, List
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def bisection_method(
    interval: Tuple[float, float], fn: Callable[[float], float], real_solution: float
) -> np.array:
    i = 0
    iter_error = []
    while interval[1] - interval[0] > 10e-6:
        a, b = interval
        m = a + (b - a) / 2
        if np.sign(fn(a)) == np.sign(fn(m)):
            a = m
        else:
            b = m
        interval = (a, b)
        iter_error.append((i, abs(real_solution - m)))
        i += 1

    return np.array(iter_error)
```

```python
fn = lambda x: np.exp(x) - 2
interval = (0, 1)
sol = np.log(2)

data = bisection_method(interval, fn, sol)
df = pd.DataFrame(data)
df.plot(x=0, y=1, kind="scatter", xlabel="Iterations", ylabel="Approximation Error", logx=True, logy=True)
```

```python
fn = lambda x: np.power(x, 3) - x - 5
interval = (0, 3)
sol = 2.094551482

data = bisection_method(interval, fn, sol)
df = pd.DataFrame(data)
df.plot(x=0, y=1, kind="scatter", xlabel="Iterations", ylabel="Approximation Error", logx=True, logy=True)
```

```python
fn = lambda x: np.power(x, 2) - np.sin(x)
interval = [0.1, np.pi]
sol = 0.876726

data = bisection_method(interval, fn, sol)
df = pd.DataFrame(data)
df.plot(x=0, y=1, kind="scatter", xlabel="Iterations", ylabel="Approximation Error", logx=True, logy=True)
```
