```python
import numpy as np
from numpy.lib.stride_tricks import as_strided
x = np.random.randint(0, 10, (3, 3, 6, 6))
N, C, H, W = x.shape
kH, kW = 2, 2
sH, sW = 2, 2
OH = ((H - kH) // sH) + 1
OW = ((W - kW) // sW) + 1

NS, CS, HS, WS = x.strides

x_s = as_strided(x, shape=(N, C, OH, OW, kH, kW), strides=(NS, CS, HS * sH, WS * sW, HS, WS), writeable=False)
```

```python
x_sr = x_s.reshape(x_s.shape[:-2] + (-1,))
a_idx = np.argmax(x_sr, axis=-1)
idx_w = np.unravel_index(a_idx, x_s.shape[-2:])
ij = np.ix_(*[np.arange(i) for i in x_s.shape[:-2]])
idx = ij + idx_w
# why not use np.zeros_like(x_s) and 
dx = np.zeros_like(x_s)
dx[idx] = 1
x
```

```python
dx.shape
```

```python
dx_u = as_strided(dx, shape=x.shape, strides=x.strides, writeable=False)
dx_u
```

```python
NS, CS, HW, WS = dout.strides
dout_p = as_strided(dout, shape=x_s.shape, strides=(NS, CS, HS * sH, WS * sW, 0, 0), writeable=True)
dout_p = dout_p.copy()
dout_p
# dout_p = as_strided(dout_p, shape=x.shape, strides=x.strides, writeable=False)
# dout_p
```
